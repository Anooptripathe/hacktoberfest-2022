```python
## importing lib.
import pandas as pd
import pandas_datareader as pdr
from datetime import datetime
```

```python
## getting the price of the shares of microsoft
df_microsoft=pdr.get_data_yahoo('MSFT') ## The dataset is of 5 years
```

```python
df_microsoft.head()
```

```python
df_microsoft['High'].plot(figsize = (12,4))
```

```python
## to get a better clear picture
df_microsoft['High'].plot(xlim= ['2019-01-01','2022-03-01'],figsize = (12,4))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
from datetime import datetime
```

```python
df_microsoft.resample(rule = 'A').max()
## hence helping us see that the max high price was in the year 2021
```

```python
df_microsoft.resample(rule = 'QS').max()['High'].plot()
# Hence helping us see that the max price was in q3 of year 2021
```

```python
df_microsoft['High'].resample(rule = 'M').mean().plot(kind='bar', figsize=(15,6))
## 2021-11-30 was the day of the heighest share price
```

### Hence the month of max Share price is 2020-11

```python
df_microsoft.head()
```

```python
df_microsoft['Low'].plot(xlim= ['2017-09-05','2021-01-01'],figsize=(12,4))
```

```python
df_microsoft.resample(rule='A').min()
#hence the min value of the share was in the year 2017 
```

```python
df_microsoft.resample(rule='BQS').min()
```

```python
df_microsoft['Low'].resample(rule = 'M').mean().plot(kind='bar', figsize=(15,6))
```

### Hence the month of lowest share price is 2017-09

```python
df_microsoft['High: 30 days rolling']=df_microsoft['High'].rolling(30).mean()
df_microsoft['Low: 30 days rolling']=df_microsoft['Low'].rolling(30).mean()
```

```python
df_microsoft.head(31)
```

```python
## for smooth trend line for the high price and low price
df_microsoft[['High: 30 days rolling','Low: 30 days rolling']].plot(figsize = (12,5))
```
